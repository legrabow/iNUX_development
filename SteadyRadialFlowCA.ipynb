# <center><ins> *Steady Radial Flow in a Confined Aquifer* <ins>

Confined aquifers are pressurized underground layers of permeable rock or sediment wedged between impermeable layers. 
**Steady radial flow** occurs when groundwater moves in a radial pattern from a central point within such an aquifer. 
This phenomenon has been of great importance throughout history, with ancient civilizations developing methods to harness and control groundwater resources.

Modern hydrogeology has refined the understanding of steady radial flow through **mathematical modeling**. 
By analyzing parameters like hydraulic conductivity and aquifer thickness, hydrogeologists can simulate and predict confined aquifer behavior. 

This knowledge has practical implications in designing and managing groundwater extraction systems for sustainable utilization.
Studying steady radial flow also helps estimate groundwater availability and replenishment, aiding water resource planning. 
It contributes to environmental conservation by assessing the impact of groundwater extraction on ecosystems and evaluating the interaction between confined aquifers and surface water bodies.

The **state of equilibrium water level** in steady radial flow in a confined aquifer represents a stable condition where *groundwater recharge equals extraction rates*. It is a crucial parameter for assessing the sustainability of groundwater extraction. If extraction exceeds recharge, the water level gradually declines, leading to drawdown and potential adverse effects. Conversely, if recharge exceeds extraction, the water level may rise, causing artesian flow. Determining the equilibrium water level involves studying hydraulic properties and using mathematical models for sustainable groundwater management.

<div class="alert alert-block alert-info">
    <b>The additional assumptions are the following :
    <br>1.  The aquifer is confined top and bottom
    <br>2.  The well is pumped at a constant rate 
    <br>3.  Equilibrium has been reached ; there is no further change in drawdown with time </b>
</div> 

From Darcy's law, the flow of water through any circular section of th aquifer toward the well is the area of the circular section times the hydraulic conductivity K, times the hydraulic gradient. 
The hydraulic gradient is expressed as the change in head with radial distance from the well, dh/dr. 
The area of a cicular section is the circumference of the aquifer, b.

Therefore, the area is *2 $\pi$ r*, where :  
    r is the radius of the circles, times the thickness of the aquifer, b.
    
Therefore, the area is *2$\pi$rb*.
Flow through the circular section can be expressed as : 
    
<center><span style="font-size: 16px;">$Q =(2\pi rb) *K * \frac{dh}{dr}  $</span></center>
where : 

Q : the pumping rate (L$^{3}$/T)

r : the radial distance from the circular section to the well (L)

b : the aquifer thickness (L)

K : the hydraulic conductivity (L/T)

dh/dr : the hydraulic gradient (dimensionless)


The equation could be rewritted because the product of the aquifer thickness and the hydraulic conductivity is the transmissivity (T) : 

<center><span style="font-size: 16px;">$Q =(2\pi r) *T * \frac{dh}{dr}  $</span></center>

<br>
In the case of a confined and horizontal flow, the equation can be rearranged to provide the following : 

<center><span style="font-size: 16px;">$dh =\frac{Q}{(2\pi T)}*\frac{dr}{r}  $</span></center>

<br> 


*<center><ins>Figure 1</ins> : Steady Radial flow through a confined aquifer : Equilibrium drawdown*

<br> 

If there are two observation wells, the head is h1 at a distance r1 from the pumping well ; it is h2 at a distance r2. 
We can integrate both sides of the previous equation with these boundary conditions : 

<br>
<center><span style="font-size: 16px;">$ \int_{h_1}^{h_2} dh  = \frac{Q}{2\pi T}*  \int_{r_1}^{r_2} \frac{dr}{r} $</span></center>
<br>
<br>
<center><span style="font-size: 16px;">$ h_2 - h_1  = \frac{Q}{2\pi T}*  \ln\frac{r_2}{r_1}  $</span></center>

<br>
<br>
<center><span style="font-size: 16px;">$ T = \frac{Q}{2\pi (h_2 - h_1)}*  \ln\frac{r_2}{r_1}  $</span></center>

<br>
Where : 

T : the aquifer transmissivity (L$^{2}$/T)

Q : pumping rate (L$^{3}$/T)

h1 : head at distance r_1 from the pumping well (L)

h2 : head at distance r_2 from the pumping well (L)


<div class="alert alert-block alert-info">
    <b> If pumping has progressed to a point where the drawdown  has stabilized, the head is determined in two bservation wells, located different distances from the pumping well.
    The pumping rate, the heads, and the radial distances are entered into the previous equation and it allows to determine the aquifer transmissivity.
    Notice that there is no provision for the aquifer storatitivity in this equation. 
    Under steady-state conditions, there is no change in head with time and water is not coming from storage.
    Hence, the Thiem equation cannot be used to determine aquifer storativity.</b>
</div>

**`PROBLEM`**

A pumping well located in a confined aquifer is being operated with a pumping rate of 220 gallons per minute. 
By measuring the *drawdown in two observation wells*, it has been determined that after 1270 min of pumping, no further drawdown is observed. 
The first observation well, referred to as Well 1, is situated at a distance of 26 feet from the pumping well and has a head of 29.34 feet above the top of the aquifer. 
The second observation well, Well 2, is located 73 feet away from the pumping well and has a head of 32.56 feet above the top of the aquifer.
To find the **aquifer transmissivity**, use the *Thiem equation*.

**`SOLUTION`**
You have to first convert the pumping rate of 220 gal/min to the equivalent rate in cubic feet per day. 
You make the conversion, even though steady-state conditions were reached, before a full day (1440min)of pumping occured. 

<center><span style="font-size: 14px;">$$ 220 \frac{gal} {min} * \frac{1~ ft^{3}} {7.48~ gal} * 1440 \frac{min} {d} = 42, 400 ~ft^{3}/d  $$</span></center> 

Now, the equation of the transmissivity, T depending on the two observation wells, could be used : 
<br>
<br>
<center><span style="font-size: 16px;">$ T = \frac{Q}{2\pi (h_2 - h_1)}*  \ln(\frac{r_2}{r_1})  $</span></center>
<br>
<br>
<center><span style="font-size: 16px;">$ T = \frac{42,400 ft^{3}}{2\pi (32.56 ~ft - 29.34 ~ft)}*  \ln(\frac{73 ~ft}{26 ~ft})  $</span></center>
<br>
<br>
<center><span style="font-size: 14px;">$ T = \frac{42,400}{20.2} *  \ln(2.81) ~ft^{2}/d $</span></center>
<br>
<br>
<center><span style="font-size: 14px;">$ T = 2710 ~ft^{2}/d  ~~~~or~~~~  T = 826~m^{2}/d $</span></center>


```python
# Importing librairies
import numpy as np
import plotly.graph_objects as go

# Define the specific flow and transmissivity
def calculate_specific_flow(Q, h1, h2, r1, r2):
    transmissivity = Q * 2 * np.pi * (h2 - h1) * np.log(r2 / r1)
    specific_flow = Q - transmissivity
    return specific_flow

# Define the parameters
Q = 1000  # Pumping rate (m^3/d)
h1 = 0.83  # Head at distance r1 from pumping well (L)
h2 = 0.92  # Head at distance r2 from pumping well (L)
r1 = 2.1   # Distance from pumping well for point 1 (r1)
r2 = 0.7  # Distance from pumping well for point 2 (r2)

# Calculation of the specific flow 
specific_flow = calculate_specific_flow(Q, h1, h2, r1, r2)
print("Specific Flow Rate (m^2/day) : Qspe = ", specific_flow)

# Define the pump cylinder
pump_radius = 0.3  # 60cm vertical diameter cylinder
pump = go.Scatter(x=[-pump_radius, pump_radius, pump_radius, -pump_radius, -pump_radius],
                  y=[h1 + pump_radius, h1 + pump_radius, -pump_radius, -pump_radius, h1 + pump_radius],
                  mode='lines',
                  line=dict(color='gray', width=3),
                  fill='toself',
                  fillcolor='gray',
                  name='Pump')

# Define the piezometric level
theta = np.linspace(0, 2 * np.pi, 100)
piezometric_level_r = r2 + 3  # Slightly outside the dynamic level
piezometric_level_x = piezometric_level_r * np.cos(theta)
piezometric_level_y = np.ones_like(theta) * h1
piezometric_level = go.Scatter(x=piezometric_level_x,
                               y=piezometric_level_y,
                               mode='lines',
                               line=dict(color='blue', width=2),
                               name='Piezometric Level')

# Define the dynamic level
dynamic_level_x = np.linspace(-5, -pump_radius, 100)
dynamic_level_y = np.linspace(h1, h1 - (h2 - h1), 100)
dynamic_level_x = np.concatenate((dynamic_level_x, dynamic_level_x[::-1]))
dynamic_level_y = np.concatenate((dynamic_level_y, dynamic_level_y[::-1]))
dynamic_level = go.Scatter(x=dynamic_level_x,
                           y=dynamic_level_y,
                           mode='lines',
                           line=dict(color='red', width=2),
                           fill='toself',
                           fillcolor='lightblue',
                           name='Dynamic Level')

# Define the mirrored dynamic level
mirrored_dynamic_level_x = -dynamic_level_x
mirrored_dynamic_level_y = dynamic_level_y
mirrored_dynamic_level = go.Scatter(x=mirrored_dynamic_level_x,
                                   y=mirrored_dynamic_level_y,
                                   mode='lines',
                                   line=dict(color='red', width=2),
                                   fill='toself',
                                   fillcolor='lightblue',
                                   name='Mirrored Dynamic Level')

# Define the segment joining the two dynamic lines
segment_x = [-pump_radius, pump_radius]
segment_y = [dynamic_level_y[np.abs(dynamic_level_x - pump_radius).argmin()], mirrored_dynamic_level_y[np.abs(dynamic_level_x - pump_radius).argmin()]]
segment = go.Scatter(x=segment_x,
                     y=segment_y,
                     mode='lines',
                     line=dict(color='red', width=2),
                     name='Segment')

# Create the merged dynamic line
merged_dynamic_line_x = np.concatenate((dynamic_level_x, segment_x, mirrored_dynamic_level_x))
merged_dynamic_line_y = np.concatenate((dynamic_level_y, segment_y, mirrored_dynamic_level_y))

# Create the merged dynamic line Scatter object
merged_dynamic_line = go.Scatter(
    x=merged_dynamic_line_x,
    y=merged_dynamic_line_y,
    mode='lines',
    line=dict(color='red', width=2),
    name='Equilibrium Water Level')

# Define the confined aquifer rectangle
b = 0.05 # Thickness of the confined aquifer (b)
aquifer_rectangle = go.Scatter(x=[-5, 5, 5, -5],
                               y=[0, 0, b, b],
                               mode='lines',
                               line=dict(color='green', width=2),
                               name='Confined Aquifer')

# Define the layout
layout = go.Layout(title='Steady Radial Flow in Confined Aquifer',
                   xaxis=dict(title='Distance from Pump, r (m)'),
                   yaxis=dict(title='Head, h (m)'),
                   showlegend=True,
                   legend=dict(x=0.93, y=0.05))



# Create the figure
fig = go.Figure(data=[pump, piezometric_level, merged_dynamic_line, aquifer_rectangle], layout=layout)

# Show the figure
fig.show()

```
