# <center><ins> *Nonequilibrium Radial Flow in a Confined Aquifer* <ins>

**Nonequilibrium Radial flow** in a *confined aquifer*, specifically using the **Theis method**, involves the analysis of transient groundwater conditions. Unlike steady radial flow, nonequilibrium flow considers situations where the extraction or recharge rates change over time, leading to dynamic variations in water levels. The Theis method, developed by Charles Theis, is a widely used analytical solution for assessing such transient flow.

In the Theis method, the aquifer is characterized by hydraulic properties such as *transmissivity* and *storativity*. By monitoring changes in water level over time at observation wells, hydrogeologists can determine the response of the aquifer to pumping or recharge activities. The Theis equation, derived from mathematical derivations, allows estimation of parameters such as *hydraulic conductivity* and *specific storage*.

The method assumes a *homogeneous and isotropic aquifer* with a **constant pumping rate or recharge rate**. By analyzing the time-drawdown relationship, hydrogeologists can infer important information about the aquifer's characteristics, such as its hydraulic conductivity and storage coefficient. This information aids in understanding the aquifer's response to transient conditions and helps in groundwater resource management.

The Theis method provides valuable insights into aquifer behavior, allowing for the prediction of future water level changes and the evaluation of pumping or recharge scenarios. It is widely utilized in practical applications such as well design, aquifer performance assessments, and estimating sustainable extraction rates.

    <b>The additional assumptions are the following :
    <br>1.  The aquifer is confined top and bottom
    <br>2.  The well is pumped at a constant rate 
    <br>3.  Equilibrium has not been reached ; there is change in drawdown with time associated to a cone of depression, growing with time </b>
</div> 

From Darcy's law, the flow through the circular section can be expressed as : 
    
<center><span style="font-size: 16px;">$Q =(2\pi r) *T * \frac{dh}{dr}  $</span></center>
where : 

Q : the pumping rate (L$^{3}$/T)

r : the radial distance from the circular section to the well (L)

T : the aquifer transmissivity (L$^{2}$/T)

dh/dr : the hydraulic gradient (dimensionless)


The equation could be rewritted as follows : 

<center><span style="font-size: 16px;">$ r \frac{\partial h}{\partial r} =  \frac{Q}{2\pi T}  $</span></center>

<br>


*<center><ins>Figure 1</ins> : Fully penetrating well pumping from a confined aquifer*

<br> 
The initial condition of a horizontal potentiometric can be expressed as :

<center><span style="font-size: 14px;"> For all r, $h(r;0)=h_0 $</span></center>

<br> 

The boundary condition signifying an infinite horizontal extent with no drawdown at any time is : 

<br>
<center><span style="font-size: 14px;"> For all t, $h(\infty;t)=h_0 $</span></center>

<br>
Under the initial and boundary conditions, the **Theis or nonequilibrium equation** is obtained : 
<br>
<center><span style="font-size: 16px;">$ h_0-h = \frac{Q}{4\pi T}  \int_{u}^{\infty}\frac{e^{-a}}{a}da  $</span></center>
<br>

where the argument u is given by : 
<center><span style="font-size: 16px;">$ u = \frac{r^{2} S}{4 t T}  $</span></center>

<br>
where : 
    
Q : the constant pumping rate (L$^{3}$/T)
    
h : the hydraulic head (L)
    
$h_0$ : the initial hydraulic head (L)
    
$h_0$ -h : the drawdown (L)

r : the radial distance from the pumping well (L)
    
t : the time since pumping began (T)
    
S : the aquifer storativity (dimensionless)

T : the aquifer transmissivity (L$^{2}$/T)

<div class="alert alert-block alert-info">
    <b> Q is a pumping rate in cubic meter or cubic feet per day. 
        Even if the well is pumped for less than 24 hours, the rate of Q must still be expressed in terms of the volume that would be pumped in a day.</b>
</div>

The integral is called **the exponential integral**. 
It can be approximated by an infinite series so the Theis equation becomes : 
    
<center><span style="font-size: 16px;">$ h_0-h = \frac{Q}{4\pi T} \Bigg[-0.5772- \ln{u} + u - \frac{u^{2}}{2.2!} + \frac{u^{3}}{3.3!} - \frac{u^{4}}{4.4!} +...\Bigg]  $</span></center>



<div class="alert alert-block alert-info">
    <b> The infinite series term has been called the well function and is generally designed as W(u).
    Although the series is easy to evaluate  mathematically, it has been conveniently tabulated (*Wenzel 1942*).
    Moreover, Appendix 1 contains a table of W(u) for various values of u. 
    </div>

    Using well function notation, the Theis equation is also rxpressed as :
        
<center><span style="font-size: 16px;">$ h_0-h = \frac{Q}{4\pi T} W(u)  $</span></center>

In this way, the Theis equation can be recast to a different form : 

<center><span style="font-size: 16px;">$ T = \frac{Q}{4\pi (h_0-h)} W(u)  $</span></center>

Where :
<center><span style="font-size: 16px;">$ S = \frac{4 t T u}{r^{2}}  $</span></center>
